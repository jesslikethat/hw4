```python
ninth = merge_school_summary.loc[merge_school_summary["grade"] == "9th"].groupby("school_name", as_index = False)
tenth = merge_school_summary.loc[merge_school_summary["grade"] == "10th"].groupby("school_name", as_index = False)
eleventh = merge_school_summary.loc[merge_school_summary["grade"] == "11th"].groupby("school_name", as_index = False)
twelfth = merge_school_summary.loc[merge_school_summary["grade"] == "12th"].groupby("school_name", as_index = False)
```

```python
ninth_math_avg_df = pd.DataFrame(round(ninth["math_score"].mean(), 1))
tenth_math_avg_df = pd.DataFrame(round(tenth["math_score"].mean(), 1))
eleventh_math_avg_df = pd.DataFrame(round(eleventh["math_score"].mean(), 1))
twelfth_math_avg_df = pd.DataFrame(round(twelfth["math_score"].mean(), 1))
```

```python
#Average Math Score by grade
avg_math_bygrade_df = pd.merge(ninth_math_avg_df, tenth_math_avg_df, on = "school_name", how = "inner")
avg_math_bygrade_df = pd.merge(avg_math_bygrade_df, eleventh_math_avg_df, on = "school_name", how = "inner")
avg_math_bygrade_df = pd.merge(avg_math_bygrade_df, twelfth_math_avg_df, on = "school_name", how = "inner")
avg_math_bygrade_df.columns = ["school name", "9th", "10th", "11th", "12th"]
avg_math_bygrade_df
```

```python

ninth_reading_avg_df = pd.DataFrame(round(ninth["reading_score"].mean(), 1))
tenth_reading_avg_df = pd.DataFrame(round(tenth["reading_score"].mean(), 1))
eleventh_reading_avg_df = pd.DataFrame(round(eleventh["reading_score"].mean(), 1))
twelfth_reading_avg_df = pd.DataFrame(round(twelfth["reading_score"].mean(), 1))
```

```python
#Average Reading score by Grad
avg_reading_bygrade_df = pd.merge(ninth_reading_avg_df, tenth_reading_avg_df, on = "school_name", how = "inner")
avg_reading_bygrade_df = pd.merge(avg_reading_bygrade_df, eleventh_reading_avg_df, on = "school_name", how = "inner")
avg_reading_bygrade_df = pd.merge(avg_reading_bygrade_df, twelfth_reading_avg_df, on = "school_name", how = "inner")
avg_reading_bygrade_df.columns = ["school name", "9th", "10th", "11th", "12th"]
avg_reading_bygrade_df
```
